## 2. Feature Engineering

### 2.1. Overview

```python
feature_columns = list(data.loc[:, 'v000':'v999'])
feature_columns
```

### 2.2. Basic Feature Selection

```python
train_features = gbx.select(
    train,
    train_statistics,
    feature_columns=feature_columns,
    thres_pctisna=95,
    thres_nunique=2,
    as_pandas=True,
)
train_features
```

```python
val_features = gbx.select(
    val,
    val_statistics,
    feature_columns=feature_columns,
    thres_pctisna=95,
    thres_nunique=2,
    as_pandas=True,
)
val_features
```

```python
test_features = gbx.select(
    test,
    test_statistics,
    feature_columns=feature_columns,
    thres_pctisna=95,
    thres_nunique=2,
    as_pandas=True,
)
test_features
```

```python
features = gbx.select(
    data,
    statistics,
    feature_columns=feature_columns,
    thres_pctisna=95,
    thres_nunique=2,
    as_pandas=True,
)
features
```

## 3. Modeling

```python
trainloader = gbx.dataloader('xgb', train[feature_columns], label=train['DEFAULT_LABEL'])
valloader = gbx.dataloader('xgb', val[feature_columns], label=val['DEFAULT_LABEL'])
testloader = gbx.dataloader('xgb', test[feature_columns])

params = {
    'booster': 'gbtree',
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'gamma': 1,
    'min_child_weight': 1.5,
    'max_depth': 5,
    'lambda': 10,
    'subsample': 0.7,
    'colsample_bytree': 0.7,
    'colsample_bylevel': 0.7,
    'eta': 0.04,
    'tree_method': 'exact',
    'seed': 2020,
    'nthread': 36,
    "silent": True,
}

model = gbx.train(
    'xgb',
    params,
    trainloader,
    num_boost_round=50000,
    evals=[(trainloader, 'train'), (valloader, 'eval')],
    early_stopping_rounds=200,
    verbose_eval=200,
)
```

```python
model.predict(valloader, ntree_limit=model.best_ntree_limit)
```

```python
trainloader = gbx.dataloader('lgb', train[feature_columns], label=train['DEFAULT_LABEL'])
valloader = gbx.dataloader('lgb', val[feature_columns], label=val['DEFAULT_LABEL'])
testloader = gbx.dataloader('lgb', test[feature_columns])

params = {
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': 'auc',
    'min_child_weight': 5,
    'num_leaves': 2**5,
    'lambda_l2': 10,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 4,
    'learning_rate': 0.1,
    'seed': 2020,
    'nthread': 28,
    'n_jobs': 24,
}

model = gbx.train(
    'lgb',
    params,
    trainloader,
    num_boost_round=50000,
    evals=[trainloader, valloader],
    early_stopping_rounds=200,
    verbose_eval=200,
)
```

```python
pred = model.predict(val[feature_columns], num_iteration=model.best_iteration)
```

```python
metrics.roc_auc_score(val['DEFAULT_LABEL'], pred)
```
